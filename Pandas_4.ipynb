#Missing Valuelarla Yapılabilecek 3 şey vardır

```python
# 1) Dokunmadan bırakabiliriz.
# 2) Drop edebiliriz (satır veya sütun bazlı)
# 3) Mantıklı bir şekilde doldurabiliriz.
```

```python
df
```

```python
# 1.Yol : Yapılacak hiçbir şey yok. Bu şekilde işlemlerine devam et.
```

```python
# 2.Yol : 
```

```python
df.dropna()
```

```python
df.dropna(how = 'any') # Herhangi bir nan varsa satırı düşürür.
```

```python
df.dropna(how = 'all') # Bütün sütunları nan sa satırı düşürür.
```

```python
df.dropna(thresh = 1) # 1 tane bile nan ollmayan veri varsa o satırı tut.
```

```python
df.dropna(thresh = 5)
```

```python
df.dropna(axis = 1, thresh = 2)
```

```python
df.dropna(axis = 1, thresh = 4)
```

```python
# 3.Yol : Fill the data (Datayı doldur.)
```

```python
df.head()
```

```python
df.dropna(how = 'all', inplace = True)
```

```python
df.head()
```

```python
df['pre_movie_score'].fillna(9.5)
```

```python
df['pre_movie_score'].fillna(df['pre_movie_score'].mean())
```
