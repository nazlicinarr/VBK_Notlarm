```python
df = pd.read_csv('tips (1).csv')
df.head()
```

```python
# Soru : Ödenen ücreti 40' tan büyük olan satırları getirim.
df[df['total_bill'] > 40]
```

```python
# Soru : Cinsiyetleri erkek olan satırları getirin.
df[df['sex'] == 'Male']
```

```python
# Soru : Cinsiyetleri kadın olan ve bahşiş miktarı 5'ten büyük olan satırları getirin.
df.head()
```

```python
df[(df['sex'] == 'Female')& (df['tip'] > 5) ]
```

```python
# Soru : Günü pazar(Sun) olan ve ödenen miktar 30'dan büyük olan satırları getirin.
df[(df['day'] == 'Sun') & (df['total_bill'] > 30)]
```

```python
# Soru : Cumartesi (Sat) veya Pazar (Sun) günü olan satırları getirin.
df[(df['day'] == 'Sat' ) | (df['day'] == 'Sun' )]
```

```python
# Soru : Cumartesi veya Pazar veya Perşembe(Thur) günü olan satırları getirin.
df[(df['day'] == 'Sat' ) | (df['day'] == 'Sun' ) | (df['day'] == 'Thur')]
```

```python
df[df['day'].isin(['Sat', 'Sun', 'Thur'])]
```

```python
df[df['day'] != 'Sun']
```

```python
df[~(df['day'] == 'Sun')]
```
