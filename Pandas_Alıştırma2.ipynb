```python
# 1) df_tips içindeki Payment ID column'unun son 3 rakamını çekin. Bu işlemi 3 farklı yolla da yapmaya çalışın.
```

```python
df_tips = pd.read_csv('tips.csv')
df_tips
```

```python
def son3(x):
  return str(x)[-3:] #1.YOL
```

```python
df_tips['Payment ID'] = df_tips['Payment ID'].apply(son3)
df_tips.head()
```

```python
df_tips['Payment ID'] = df_tips['Payment ID'].apply(lambda x : str(x)[-3:] ) #2.YOL
df_tips.head()
```

```python
df_tips['Payment ID'].astype(str).str[-3:] #3.YOL
```
